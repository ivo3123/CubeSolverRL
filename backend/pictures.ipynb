```python
import plotly.express as px
import plotly.graph_objects as go
import polars as pl
from polars import col as c
```

```python
pl_data = pl.read_parquet("docs/1_edges_first_layer_v1/training_data.parquet")

pl_data
```

```python
fig = go.Figure()

x = pl_data["episode"]
y1 = pl_data["edges_solved_mean"]
y2 = pl_data["epsilon"]

fig.add_trace(go.Scatter(
    x=x, y=y1,
    mode='lines',
    name='Edges solved rolling 100 mean'
))

fig.add_trace(go.Scatter(
    x=x, y=y2,
    mode='lines',
    name='Epsilon',
    yaxis='y2'
))

fig.update_layout(
    title={
        'text': f'Solved edges',
        'x': 0.5,
        'y': 0.94,
    },
    xaxis=dict(title='Episode'),
    yaxis=dict(title='Solved Edges'),
    yaxis2=dict(
        title='Epsilon',
        overlaying='y',
        side='right'
    ),
    height=725,
    legend=dict(
        orientation='h',
        yanchor='bottom',
        y=-0.3,
        xanchor='center',
        x=0.5
    ),
)

fig.show()
```
